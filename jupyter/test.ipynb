```python
import requests
import time

def get_data():

    url = "http://get_data_api:8003/data"

    params = {
        "group_number": 3
    }

    for _ in range(10):
        try:
            response = requests.get(url, params=params)
            return response.json()
        except:
            print("Waiting for API...")
            time.sleep(3)
```
